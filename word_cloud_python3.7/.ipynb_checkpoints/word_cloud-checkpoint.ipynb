## 1a - Import the data <a id="1a"></a><<a href="#sub-outline_1">Shortcut To Sub-Outline</a>>

```python
# import the data into a data frame
raw_data = pd.read_csv('scraped_data/scraped_data_data-scientist_4-24-2019.txt')

# extract the job_title and state column
raw_data = raw_data[['job_title','state']]

# filter the data frame rows by the state CA
state = 'CA'
data = []
data = raw_data[raw_data['state']==state]
print(data.head()) # view the data before dropping the column

# drop the state column, now filled with the string 'CA'
data = data.drop(columns='state')

# view the data 
data.head()
```

# Does the data look like what we expected?

- the data includes **only CA job titles**


- the data includes **only the job titles column**


## 1b - Convert all Word Cloud text to a single lowercase long string <a id="1b"></a><<a href="#sub-outline_1">Shortcut To Sub-Outline</a>>

Here, we **make a function that converts our column of job titles into a single string of lowercase, space-separated text.**

```python
def get_long_string(data):
    '''
    This function transforms a column of text into a long string of space-separated,
    lowercase words.
    
    input: 
    - data: a data frame that is a column of text with the column label 'job_title'
    
    output: 
    - long_string: a single long string of lowercase words, separated by spaces 
    '''

    # we need an empty string to store all of the words
    long_string = ' '
    
    
    # iterate through each row of the job title column 
    for value in list(data['job_title']): 

        # change each value to string 
        value = str(value) 

        # split the job titles into a list separated by spaces  
        sub_words = value.split() 

        # i. converts each sub-word from the job title into lowercase 
        sub_words = [value.lower() for value in sub_words]
        
        for word in sub_words: 

            # ii. append each sub-word to the long string
            long_string = long_string + word

            # iii. separate each word by a space
            long_string = long_string + ' ' 
                
    return long_string
```

Now let's **retrieve our long string of lowercase words** and **take a look at the data**

```python
# retrieve the long lowercase string of words
long_string = get_long_string(data)

# view the data
long_string[0:1000]
```

# Does the data look like what we expected?

The data matches the **three characteristics that we were aiming to achieve:**

- each word in the data consists of **only lowercase letters**


- the data is a **single long string**

  
- the data has **each word separated by a space**
 

## 1c - Filter the stopwords and get word frequencies from the text data <a id="1c"></a><<a href="#sub-outline_1">Shortcut To Sub-Outline</a>>

```python
# First, filter out the symbols with the re.sub method
long_string = re.sub(
                    '[^a-zA-Z+#0234]', # keep a-z, A-Z, #, +, 2-4, for c#, c++, d3, h20, neo4 
    
                     ' ', # replace anything else with a space
    
                     long_string # input the long_string
                    ) 

# Second, filter out the stopwords and get the dictionary of words and word frequencies
words_and_freqs_dict = WordCloud().process_text(long_string)

# view the data
list(words_and_freqs_dict.items())[0:15]
```

# Does the data look like what we expected?

The data matches the **three characteristics that we were aiming to achieve:**
- the data has **no symbols**

  
- the data has **no stopwords**

 
- the data is **a dictionary with words and word frequencies**

## 1d - Make the word cloud <a id="1d"></a><<a href="#sub-outline_1">Shortcut To Sub-Outline</a>>

- #### i. create the word cloud object

```python
# create the word cloud object
wc1 = WordCloud(
               relative_scaling = .5, # change the size of the word based on frequency (try between 0.4 and 0.6)

               background_color='black', # set the image background color

               min_font_size=2, # set smallest word cuttoff

               scale=4, # Adjust word resolution in the picture

               colormap='hsv'# change color palette
               )

```

- #### ii. fit the word cloud object with the data

```python
# fit the word cloud object to the data
wc1.fit_words(words_and_freqs_dict)  
```

## 1e - <a id="1e"></a>Plot the word cloud <<a href="#sub-outline_1">Shortcut To Sub-Outline</a>>

```python
job_search = 'data-scientist'

# i. format the figure

# store the figure object
fig = plt.figure(figsize=(40,20))
    
# create a title for the figure
plt.title('Words in '+job_search+' related job titles in '+state,

          fontsize = 90, # make the figure title words big

          y = 1.01 # add space between the title and figure
         )

# ii. format the axes

# store the axis object 
ax = plt.gca()

# Make axes fill image
plt.tight_layout(pad=0)

# remove ticks from axis
ax.axis('off')

# insert the word cloud image into the axis object ax 
ax.imshow(wc1)

# iii. save the figure
plt.savefig('figs/'+'word_cloud_'+job_search+'_'+state+'.png',bbox_inches='tight')

# iv. plot the figure
plt.show()
```

##  2e - <a id="2e"></a>Recolor the word cloud <<a href="#sub-outline_2">Back To Sub-Outline</a>>

in this section we will first **create an input for the color_func attribute in the WordCloud.recolor method.**

we will then **change the color of the words in the word cloud.**

- ### i. Find the norm of the frequency values  

```python
# Get the max value of the frequency values
minimum = min(words_and_freqs_dict.values()) 

# Get the min value of the frequency values
maximum = max(words_and_freqs_dict.values())

# Get the norm of the dictionary
normal = Normalize(vmin = minimum, vmax = maximum)
```

- ### ii. Choose a cmap

```python
# we choose a color palette for the words 
color_palette = 'gist_rainbow'
```

- ### iii. Create the ScalarMapping object

```python
# Plug the norm and cmap values into the ScalarMappable parameters
sm = ScalarMappable(norm = normal, cmap = color_palette)

# Store for comparison
sm2 = sm
```

- ### iv. Create color_func 
**This function is trouble** to deal with, but **it works with close attention to two very important details...**

   First, **declare the ScalarMapping object as a global variable** before using color_func.

   Second, **declare the words_and_freqs_dist as a global variable** before using color_func.
    
   ref - color_func source code: https://amueller.github.io/word_cloud/_modules/wordcloud/wordcloud.html#WordCloud.fit_words

```python
def color_function(word, font_size, position, orientation, random_state, **kwargs):
    '''
    This function converts the frequency of a word into its rgba mapping in 0-255 
    range, and returns a rgb string to the color_func attribute in the WordCloud.recolor
    method.
    
    This funciton uses a globally defined ScalarMapping object,
    and a globally defined words_and_freqs_dict dictionary.
    
    input:
    - ref - http://amueller.github.io/word_cloud/auto_examples/a_new_hope.html
    
    output: 
    - "rgb(%.0f,%.0f,%.0f)" % (r, g, b): a string that contains the rgb value of each frequency
    '''
    #--> words_and_freqs_dict and sm must be globally declared before calling this function
    # Get the frequency of a word
    freq=words_and_freqs_dict[word] 

    #--> sm must be globally declared before calling this function
    # Get the rgba_value of a frequency in the 0-255 value range
    rgba_value = sm.to_rgba(
                           freq, # the freq value of a word
                           bytes = True # Convert to 0-255 color value
                          )
    
    # extract the red part
    r = rgba_value[0]
    
    # extract the green part
    g = rgba_value[1]
    
    # extract the blue part
    b = rgba_value[2]

    return "rgb(%.0f,%.0f,%.0f)" % (r, g, b)
```

- ### v. Recolor the word cloud

```python
# Use the WordCloud().recolor() method with the color_function
wc2.recolor(color_func = color_function)
```

##  3f - <a id="3f"></a>Recolor the word cloud <<a href="#sub-outline_3">Back To Sub-Outline</a>>

in this section we will first **create an input for the color_func attribute in the WordCloud.recolor method.**

we will then **change the color of the words in the word cloud.**

- ### i. Find the norm of the log(frequency) values  

```python
# Get the max value of the log(frequency) values
minimum = min(words_and_freqs_dict.values()) 

# Get the min value of the log(frequency) values
maximum = max(words_and_freqs_dict.values())

# Get the norm of the dictionary
normal = Normalize(vmin = minimum, vmax = maximum)
```

- ### ii. Choose a cmap

```python
# we choose a color palette for the words 
color_palette = 'gist_rainbow'
```

- ### iii. Create the ScalarMapping object

```python
# Plug the norm and cmap values into the ScalarMappable parameters
sm = ScalarMappable(norm = normal, cmap = color_palette)

# Store for comparison
sm3 = sm
```

- ### iv. Create color_func 
**This function is trouble** to deal with, but **it works with close attention to two very important details...**

   First, **declare the ScalarMapping object as a global variable** before using color_func.

   Second, **declare the words_and_freqs_dist as a global variable** before using color_func.
    
   ref - color_func source code: https://amueller.github.io/word_cloud/_modules/wordcloud/wordcloud.html#WordCloud.fit_words

```python
def color_function(word, font_size, position, orientation, random_state, **kwargs):
    '''
    This function converts the log(frequency) of a word into its rgba mapping in 0-255 
    range, and returns a rgb string to the color_func attribute in the WordCloud.recolor
    method.
    
    This funciton uses a globally defined ScalarMapping object,
    and a globally defined words_and_freqs_dict dictionary.
    
    input:
    - ref - http://amueller.github.io/word_cloud/auto_examples/a_new_hope.html
    
    output: 
    - "rgb(%.0f,%.0f,%.0f)" % (r, g, b): a string that contains the rgb value of each log(frequency)
    '''
    #--> words_and_freqs_dict and sm must be globally declared before calling this function
    # Get the log(frequency) of a word
    freq=words_and_freqs_dict[word] 

    #--> sm must be globally declared before calling this function
    # Get the rgba_value of a log(frequency) in the 0-255 value range
    rgba_value = sm.to_rgba(
                           freq, # the freq value of a word
                           bytes = True # Convert to 0-255 color value
                          )
    
    # extract the red part
    r = rgba_value[0]
    
    # extract the green part
    g = rgba_value[1]
    
    # extract the blue part
    b = rgba_value[2]

    return "rgb(%.0f,%.0f,%.0f)" % (r, g, b)
```

- ### v. Recolor the word cloud

```python
# Use the WordCloud().recolor() method with the color_function
wc3.recolor(color_func = color_function)
```

# Make a Function to Get Word Clouds for all the States with Masks (with Only the Essentials )

```python
def get_word_cloud_with_mask(state,small_scale,scale_factor,raw_data,job_search):
    '''
    This function calculates the word cloud for job titles in a state and uses a mask to 
    make the word cloud in the shape of the state.
    
    input:
    - state: the state that the word cloud is being made for  
    
    - small_scale: the scale of the small scale image
    
    - scale_factor: the scale factor that multiplies the scalle of he small scale image to
                    get the full scale image
    
    - raw_data: the job titles and state names for every state
    
    - job_search: the job title used to search for the data
    
    output: 
    There is no output of the function. A large scale word cloud image is saved for every 
    state in the us_states_word_clouds folder.
    
    '''
    try: 
        # filter the data frame rows by the state
        data = []
        data = raw_data[raw_data['state']==state]

        # drop the state column, now filled with the string of the selected state
        data = data.drop(columns='state')

        # get the long string
        long_string = get_long_string(data)

        # filter out the symbols from the long string with the re.sub method
        long_string = re.sub(
                            '[^a-zA-Z+#0234]', # keep a-z, A-Z, #, +, 2-4, for c#, c++, d3, h20, neo4 

                             ' ', # replace anything else with a space

                             long_string # filter long_string
                            ) 

        # filter out the stopwords from the long string and get the word/word-freq dictionary
        words_and_freqs_dict = WordCloud().process_text(long_string)

        # store the mask image as greyscale
        img = cv2.imread('us_states_masks/'+str(state)+'_scale_'+str(small_scale)+'_mask.png',0)

        # invert the mask image for wc input
        img = cv2.bitwise_not(img)

        ## WC takes mask as 2D numpy array
        state_mask = np.array(img)

        # create the word cloud
        wc = WordCloud(background_color="black",
                       mask=state_mask,
                       contour_width=5,
                       colormap='gist_rainbow',
                       relative_scaling = .5,
                       scale=scale_factor,
                       min_font_size=2,
                       contour_color='steelblue')

        # generate word cloud
        wc.fit_words(words_and_freqs_dict)

        # Save Wordcloud as image
        wc.to_file('us_states_word_clouds/'+'masked_word_cloud_'+str(state)+'_scale_'+str(small_scale*scale_factor)+'.png')
        
        # Print a success statement when wc is made successfully
        print('Word cloud was made successfully for: '+str(state))
    # Sometimes there is no data for a given state, so we just print the error for this case
    except Exception as e:
        print(str(state)+' has error: '+str(e))
```

# Convert all Word Cloud text to a single lowercase long string 

```python
def get_long_string(data):
    '''
    This function transforms a column of text into a long string of space-separated,
    lowercase words.
    
    input: 
    - data: a data frame that is a column of text with the column label 'job_title'
    
    output: 
    - long_string: a single long string of lowercase words, separated by spaces 
    '''

    # we need an empty string to store all of the words
    long_string = ' '
    
    
    # iterate through each row of the job title column 
    for value in list(data['job_title']): 

        # change each value to string 
        value = str(value) 

        # split the job titles into a list separated by spaces  
        sub_words = value.split() 

        # i. converts each sub-word from the job title into lowercase 
        sub_words = [value.lower() for value in sub_words]
        
        for word in sub_words: 

            # ii. append each sub-word to the long string
            long_string = long_string + word

            # iii. separate each word by a space
            long_string = long_string + ' ' 
                
    return long_string
```
