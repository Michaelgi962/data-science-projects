```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
import random
```

# Import data

```python
# Get the data scientist job skills table
data = []
data = pd.read_csv('scraped_data/scraped_data_data-scientist_4-24-2019.txt')
skills = []
skills = list(pd.read_csv('skill_lists/data_scientist.txt'))
# We currently dont have data on 'pydoop', 'dumbo', 'api', 'hadoopy', or 'mrjob' :(
skills = [x for x in skills if x not in['pydoop', 'dumbo', 'api', 'hadoopy', 'mrjob']]
```

# Filter the data 

```python
# The states with the most data scientist related positons will have jobs that 
# are more closely related to data science
most_populated_states_and_totals = []
most_populated_states_and_totals = data[['state','total_jobs']].groupby('state').mean().sort_values(by='total_jobs',ascending=False)
```

```python
# The top 9 states have over 1500 jobs
most_populated_states_keys = []
most_populated_states_keys = list(most_populated_states_and_totals.index[0:9])
```

```python
# Get the dataframe only with states in the most populated state list
data_filtered = []
data_filtered = data[[(x in most_populated_states_keys) for x in data['state']]]
```

# Bin the data

```python
bin_size = int(np.sqrt(len(data_filtered)))
# Assign bin numbers
bin_assignment = []
bin_assignment = [int(i/bin_size)+1 for i in range(len(data_filtered))]
# Assign random numbers
rand_nums = []
rand_nums = [random.uniform(0,1) for i in range(len(data_filtered))]
# Sort in dataframe
rand_bin_assignment = []
rand_bin_assignment = pd.DataFrame(bin_assignment,columns = ['bin_assignment'])
rand_bin_assignment['rand_nums'] = rand_nums
rand_bin_assignment = rand_bin_assignment.sort_values(by='rand_nums')
print(rand_bin_assignment.head(2))
# Add the bin assignment column to the original dataframe
data_filtered['bin_assignment'] = list(rand_bin_assignment['bin_assignment'])
data_filtered.head(2)
# Get the averages of each group of bins
data_binned_avgs = data_filtered.groupby('bin_assignment').mean()[skills]
```

# Filter the data

```python
# There are many empty columns
temp = np.round(data_binned_avgs.mean(),decimals=2) # round the mean value of the skills to 3 decimals 
non_zero_keys = [key for key in temp.keys() if temp[key] != 0]
data_binned_avgs = data_binned_avgs[non_zero_keys]
```

```python
plt.figure(figsize=(10,10))
data_binned_avgs.mean().sort_values(ascending=False).plot(kind='bar')
```

# Scale the data
- ## Standardize the data to have an average value of zero and a variance of 1 using the formula:

# $$x' = \frac{x-mean(x)}{std(x)}$$

```python
from sklearn.preprocessing import StandardScaler
```

```python
# The data needs to be standardized for PCA
scalar = StandardScaler()

# Fit the data
scalar.fit(data_binned_avgs)

# Transform the data
data_binned_avgs_standardized = []
data_binned_avgs_standardized = scalar.transform(data_binned_avgs)
data_binned_avgs_standardized = pd.DataFrame(data = data_binned_avgs_standardized,
                                             columns = non_zero_keys)
data_binned_avgs_standardized.head(3)
```

# Get the Correlation Matrix

```python
corr_matrix = []
corr_matrix = np.round(data_binned_avgs_standardized.corr(),decimals=4)
corr_matrix.head(3)
```

```python
eigenValues_trim = eigenValues
tot = sum(eigenValues_trim)
var_exp = [(i / tot)*100 for i in sorted(eigenValues_trim, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

trace1 = dict(
    type='scatter',
    x = sorted_by_eig_val_titles[::-1],
    #x=['PC %s' %i for i in range(1,len(eigenValues)+1)],
    y=var_exp,
    name='Individual'
)

trace2 = dict(
    type='scatter',
    x = sorted_by_eig_val_titles[::-1],
    # x=['PC %s' %i for i in range(1,len(eigenValues)+1)], 
    y=cum_var_exp,
    name='Cumulative'
)

data = [trace1, trace2]

layout=dict(
    title='Explained variance by different principal components',
    yaxis=dict(
        title='Explained variance in percent'
    ),
    annotations=list([
        dict(
            x=1.16,
            y=1.05,
            xref='paper',
            yref='paper',
            text='Explained Variance',
            showarrow=False,
        )
    ])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='selecting-principal-components')
```

# K-means Clustering

```python
from sklearn.cluster import KMeans
# set up the data as an array 
data_kmeans = np.array(data_binned_avgs_standardized_pca)    
inertia = []

def get_kmeans_inertia(num_clusters,data_kmeans):
    '''
    This function gets the inertia for a k-means cluster.
    
    num_clusters: the number of clusters 
    data_kmeans: the input data for the k-means algorithms
    
    '''
    # Number of clusters
    kmeans = KMeans(n_clusters=num_clusters,n_init=100)
    # Fitting the input data
    kmeans = kmeans.fit(data_kmeans)
    # Get inertia
    return kmeans.inertia_
```

```python
# We need to find the optimum cluster size
for n in np.arange(2,11):
    inertia.append(get_kmeans_inertia(n,data_kmeans))
```

```python
plt.figure(figsize=(15,15))
plt.plot(np.arange(2,11),inertia)
plt.show()
```

# $$ The\ elbow\ is\ at\ k=3$$

# Get cluster labels for 3 clusters

```python
def get_kmeans_labels(num_clusters,data_kmeans):
    '''
    This function gets the labels for a k-means cluster.
    
    num_clusters: the number of clusters 
    data_kmeans: the input data for the k-means algorithms
    
    '''
    # Number of clusters
    kmeans = KMeans(n_clusters=num_clusters,n_init=100)
    # Fitting the input data
    kmeans = kmeans.fit(data_kmeans)
    # Getting the cluster labels
    labels = kmeans.predict(data_kmeans)
    # Get labels
    return labels
```

```python
# We need the labels to attatch to the table
labels = get_kmeans_labels(3,data_kmeans)
```

# Make the word clouds

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS) 
```

# Convert the Word Cloud Text into a single string

```python
def get_comment_words(data):
    comment_words = ' '
    
    # iterate through the job title list 
    for val in list(data): 

        # typecaste each val to string 
        val = str(val) 

        # split the title  
        sub_words = val.split() 

        # Converts each sub-word into lowercase 
        for word in range(len(sub_words)): 
            sub_words[word] = sub_words[word].lower() 

        for word in sub_words: 
              comment_words = comment_words + word + ' '
    return comment_words
```

# Make the word cloud

```python
def make_cloud(comment_words,rel_scale):
    '''
    comment_words : the sting of words to be inserted into the cloud 
    rel_scale: the scale of the size of the word as a function of frequency 
    '''
    
    #Convert all the required text into a single string here 
    #and store them in word_string

    #you can specify fonts, stopwords, background color and other options
    wordcloud = WordCloud(stopwords=stopwords,
                              background_color='black',
                              # Change smallest word
                              min_font_size=2,
                              # Increase word resolution
                              scale=20,
                              # Change the size of the word based on frequency (try between 0.4 and 0.6)
                              relative_scaling = rel_scale,
                              # Change color pallet
                              colormap='hsv'
                             ).generate(comment_words)

    plt.figure(figsize=(40,20))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

# Word Cloud First Cluster

```python
comment_words = get_comment_words(df0_titles)
make_cloud(comment_words,.5)
```

```python
comment_words = get_comment_words(df1_titles)
make_cloud(comment_words,.5)
```
